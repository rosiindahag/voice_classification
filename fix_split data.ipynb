```python
#normalisasi
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df.iloc[:,1:number_of_features] = scaler.fit_transform(df.iloc[:,1:number_of_features])
df
```

### Split 80% train data and 20% test data

```python
#random dataset
X = np.array(df)
np.random.shuffle(X)
#X = X.sample(frac = 1) untuk data yang berbentuk dataframe
print(X)
```

### Notes:
###### index 0 = musicId,
###### index 1 to 52 = features, 
###### index 53: arousal(mean), index 54: valence(mean)

```python
#split data 80% for data training and 20% for data testing
train_data = X[:int(len(X)*.80+1)] #Remaining 80% to training set
test_data = test_data = X[int(len(X)*.80+1):] #Splits 20% data to test dataset
total_train = np.size(train_data,axis=0)
total_test = np.size(test_data,axis=0)
print("training data: "+str(total_train))
print("testing data: "+str(total_test))
print("total data: "+str(total_train+total_test))
```

```python
df_train = pd.DataFrame(train_data)
df_train[0] = np.int_(df_train[0])
df_train = df_train.rename({0: 'musicId', 53: 'arousal(mean)', 54:'valence(mean)'}, axis=1)
df_test = pd.DataFrame(test_data)
df_test[0] = np.int_(df_test[0])
df_test = df_test.rename({0: 'musicId', 53: 'arousal(mean)', 54:'valence(mean)'}, axis=1)
df_train
```

```python
#save_data
df_train.to_csv("data_training_80.csv", sep=',', index=False)
df_test.to_csv("data_testing_20.csv", sep=',', index=False)
```
